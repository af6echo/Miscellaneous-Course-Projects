```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("dark_background")
```

```python
def CA_generator(x, num_steps:int=100, rule:str="01001110"):
    """
    Cellular Automaton Generator with Periodic Boundary Condition

    x: generation zero (initial condition)

    num_steps: number of time steps of simulation

    rule: the cellular automaton rule given in binary (NOT DECIMAL) as a string
    """

    L = len(x)
    # mutable copy of x:
    x_new = x.copy()
    # This matrix will hold all the data and will be used to visualize the automaton:
    image_matrix = np.zeros(shape=(num_steps+1, L), dtype=int)
    image_matrix[0] = x
    # Iterating over all the places in L, and doing it for the number of time steps specified:
    for t in range(num_steps):
        for i in range(L):
            if x[i-1]==0 and x[i]==0 and x[(i+1)%L]==0:
                x_new[i] = int(rule[-1])
            elif x[i-1]==0 and x[i]==0 and x[(i+1)%L]==1:
                x_new[i] = int(rule[-2])
            elif x[i-1]==0 and x[i]==1 and x[(i+1)%L]==0:
                x_new[i] = int(rule[-3])
            elif x[i-1]==0 and x[i]==1 and x[(i+1)%L]==1:
                x_new[i] = int(rule[-4])
            elif x[i-1]==1 and x[i]==0 and x[(i+1)%L]==0:
                x_new[i] = int(rule[-5])
            elif x[i-1]==1 and x[i]==0 and x[(i+1)%L]==1:
                x_new[i] = int(rule[-6])
            elif x[i-1]==1 and x[i]==1 and x[(i+1)%L]==0:
                x_new[i] = int(rule[-7])
            elif x[i-1]==1 and x[i]==1 and x[(i+1)%L]==1:
                x_new[i] = int(rule[-8])
    

        # Putting each new generation in the matrix:
        image_matrix[t+1] = x_new
        # Replacing each new generation with the old one:
        x = x_new.copy()
    return image_matrix
```

```python
rule = 110
# Binary Representation of the Rule:
bin_rule=np.binary_repr(rule,8)

fig, ax = plt.subplots(2,2, figsize=(12, 12))
fig.suptitle(f"Cellular Automaton\n Rule: {rule}\nBinary Representation: {bin_rule}")
# Length of the base:
L = 201
# Initial Condition: middle cell is on
x = np.zeros(L)
x[(L-1)//2] = 1
image_matrix = CA_generator(x=x, num_steps=300, rule=bin_rule)
ax[0,0].imshow(image_matrix, interpolation="none", origin="lower")
ax[0,0].set_title("Generation Zero: Middle Cell")
ax[0,0].set_xlabel("System Length")
ax[0,0].set_ylabel("Time Steps")
ax[1,0].imshow(image_matrix[:50,50:110], interpolation="none", origin="lower")
ax[1,0].set_title("Zoomed in")
ax[1,0].set_xlabel("System Length")
ax[1,0].set_ylabel("Time Steps")
ax[1,0].set_xticklabels(range(40,110,10))
# Random Inital Condition:
x = np.random.randint(0,2,L)
image_matrix = CA_generator(x=x, num_steps=300, rule=bin_rule)
ax[0,1].imshow(image_matrix, interpolation="none", origin="lower")
ax[0,1].set_title("Generation Zero: Random")
ax[0,1].set_xlabel("System Length")
ax[0,1].set_ylabel("Time Steps")
ax[1,1].imshow(image_matrix[:50,50:110], interpolation="none", origin="lower")
ax[1,1].set_title("Zoomed in")
ax[1,1].set_xlabel("System Length")
ax[1,1].set_ylabel("Time Steps")
ax[1,1].set_xticklabels(range(40,110,10));
```

```python
from matplotlib import animation
from matplotlib.ticker import FixedLocator


# to save ram:
plt.close("all")

# Added the line below from stackoverflow so that jupyter notebook would work.
# Without it, the code runs into an error that I have't quite figured out what exactly it is;
# Best I understand, some encoder (FFMpeg encdoer) is not installed.
plt.rcParams["animation.html"] = "jshtml"

L = 21
num_steps = 50
rule=np.binary_repr(110,8)
# rule="01101010"
# rule="11101110"
# rule="00101110"
# Initiating the matrix:
image_matrix = CA_generator(x=np.random.randint(0,2,L), num_steps=num_steps, rule=rule)



# Initiating the plot:
fig, ax = plt.subplots(figsize=(6,6))
ax.grid()
ax.xaxis.set_major_locator(FixedLocator(range(-1,L+1)))
ax.yaxis.set_major_locator(FixedLocator(range(0,num_steps)))
ax.tick_params(axis="y", labelleft=False, left=False)
ax.set_xlim((0, L))
ax.set_ylim((-1, num_steps))
ax.set_ylabel("Time$\longrightarrow$")
ax.set_title("Cellular Automaton Rule 110")

# Create an empty line object with no data.
# It will be updated during each frame of the animation.
line, = ax.plot([], [], "s", color="aqua")

# initialization function: plot the background of each frame.
# There was a similar function in all the example codes I studied; must be the standard way!
def init():
    line.set_data([], [])
    return line,


# The animation function. This is called sequentially.
def animate(i):
    # Adjusting for the first frame that has to be an empty plot:
    if i==0:
        return line,
    i = i-1
    x = np.nonzero(image_matrix[i])[0]
    y = [i]*len(x)
    line.set_data(x, y)  # The standard way to update x and y.
    return line,


# Finally moving on to the animator function.
# blit=True means only re-draw the parts that have changed; this helps increase the speed of the program.
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=num_steps, blit=True)
```
